```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading data into a DataFrame
file1 = pd.read_csv('marketing_customer_analysis.csv')
```

```python
file1
```

```python
numerical = pd.DataFrame(file1.select_dtypes(np.number))
numerical.head(1)
```

```python
categorical = pd.DataFrame(file1.select_dtypes(object))
categorical.head(1)
```

```python
for i in numerical.columns:
    sns.displot(x = i, data = numerical)
    plt.show()
```

```python
for i in numerical.columns:
    plt.hist(x = file1[i], density = False)  # density=False would make counts
    plt.ylabel('Count')
    plt.xlabel(i)
    plt.show()
```

```python
correlations_matrix = file1.corr()
correlations_matrix =correlations_matrix.round(2)
correlations_matrix
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```
